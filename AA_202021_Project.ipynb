```python
def show_nulls(df):
    df_nulls = (df.isnull().sum()).to_frame(name = 'Nan')
    df_nulls['Count'] = len(df)
    # df_nulls['%'] = df_nulls['Nan'] / df_nulls['Count'] * 100
    df_nulls['Count'].plot(color='red')
    ax = df_nulls['Nan'].plot.bar(figsize=(30, 5))
    for p in ax.patches:
        ax.annotate(p.get_height(), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')
    plt.show()
    df_nulls = df_nulls.drop(columns=['Count'])
    return df_nulls[df_nulls['Nan'] != 0].transpose()
```

```python
def show_distribution(df, column, c='blue'):
    values = pd.value_counts(df[column]).sort_index()
    ax = values.plot.bar(figsize=(30, 5), color=c, title=column)
    values = values.to_frame().transpose()
    for p in ax.patches:
        ax.annotate(p.get_height(), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')
    plt.show()
    return values
```

```python
def show_balance(df, column):
    df_balanced = df[column].value_counts().to_frame()
    df_balanced.plot.pie(y=column, figsize=(5, 5), autopct='%1.0f%%')
```

```python
def class_check(df, columns):
    result = {}
    df_invalids = {}
    for column in columns:
        types = df[column[0]].value_counts().to_frame()
        diffs = np.setdiff1d(types.index.values, column[1])
        if len(diffs) != 0:
            result[column[0]] = diffs.tolist()
            values = show_distribution(df, column[0], ['tab:red' if value in result[column[0]] else 'tab:blue' for value in types.sort_index().index])
            values = values[diffs]
            df_invalids[column[0]] = [values.transpose().values.sum()]
            # print(values.transpose().values.sum())
    return result, pd.DataFrame(df_invalids)
```

```python
def transform_text(df, columns):
    encoders = {}
    for column in columns:
        le = LabelEncoder()
        le.fit(df[column])
        encoders[column] = le
        df[column] = le.transform(df[column])
    return df, encoders
```

```python
def inverse_transform_text(df, columns, encoders):
    for column in columns:
        df[column] = encoders[column].inverse_transform(df[column])
    return df
```

```python
def cut(df, columns):
    for column in columns:
        df[column[0]] = pd.cut(df[column[0]], bins=column[1])
    return df
```
