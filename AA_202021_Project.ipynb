Now, we have some functions that were useful to understand the data but to see some results later.

```python
def get_nulls(df, plot=False):
    df_nulls = (df.isnull().sum()).to_frame(name = 'Nan')
    df_nulls['Count'] = len(df)
    if plot:
        df_nulls['Count'].plot(color='red')
        ax = df_nulls['Nan'].plot.bar(figsize=(30, 5))
        for p in ax.patches:
            ax.annotate(p.get_height(), 
                       (p.get_x() + p.get_width() / 2., p.get_height()), 
                       ha = 'center', va = 'center', 
                       xytext = (0, 9), 
                       textcoords = 'offset points')
        plt.show()
    df_nulls = df_nulls.drop(columns=['Count'])
    return df_nulls[df_nulls['Nan'] != 0].transpose()
```

```python
def get_distribution(df, column, c='tab:blue', plot=False):
    values = pd.value_counts(df[column]).sort_index()
    if plot:
        ax = values.plot.bar(figsize=(30, 5), color=c, title=column)
        for p in ax.patches:
            ax.annotate(p.get_height(), 
                       (p.get_x() + p.get_width() / 2., p.get_height()), 
                       ha = 'center', va = 'center', 
                       xytext = (0, 9), 
                       textcoords = 'offset points')
        plt.show()
    return values.to_frame().transpose()
```

```python
def show_balance(df, column):
    df_balanced = df[column].value_counts().to_frame()
    df_balanced.plot.pie(y=column, figsize=(5, 5), autopct='%1.0f%%')
```

```python
def check_categoric_features(df, columns, plot=False):
    result = {}
    df_invalids = {}
    for column in columns:
        types = df[column[0]].value_counts().to_frame()
        diffs = np.setdiff1d(types.index.values, column[1])
        if len(diffs) != 0:
            result[column[0]] = diffs.tolist()
            values = get_distribution(df, column[0], ['tab:red' if value in result[column[0]] else 'tab:blue' for value in types.sort_index().index], plot)
            values = values[diffs]
            df_invalids[column[0]] = [values.transpose().values.sum()]
            # print(values.transpose().values.sum())
    return result, pd.DataFrame(df_invalids)
```

```python
def load_data(df, target):
    table_X = df.drop(columns=[target]).values
    table_y = df[target].values      
    features = df.drop(columns=[target]).columns.values
    target_name = df[target].to_frame().columns.values
    return table_X, table_y, features, target_name
```

Function `cut` separate numerical features with infinity possible values into a categorical one defined by ranges.

```python
def cut(df, columns):
    for column in columns:
        df[column[0]] = pd.cut(df[column[0]], bins=column[1])
    return df
```

The function `transform_text` transforms categorical features (with letters or ranges in possible values) into numbers to be possible to use that features in some methods that only support numerical categories. The function saves the encoders of each transformed attribute, for the `inverse_transform_text` function could undo the transformation.

```python
def transform_text(df, columns):
    encoders = {}
    for column in columns:
        le = LabelEncoder()
        le.fit(df[column])
        encoders[column] = le
        df[column] = le.transform(df[column])
    return df, encoders
```

```python
def inverse_transform_text(df, columns, encoders):
    for column in columns:
        df[column] = encoders[column].inverse_transform(df[column])
    return df
```

## 1.2. Learning Classifiers

```python
def cross_validation_model(donor_X, donor_y, classifier, splits, sme=SMOTE(random_state=42), plot=False, average='binary'):
    kf = KFold(n_splits=splits, shuffle=True, random_state=42)
    scores = {}
    scores['Recall Score'] = []
    scores['F1 Score'] = []
    scores['Precision Score'] = []
    scores['Accuracy Score'] = []
    scores['Accuracy Score (train)'] = []
    
    for fold, (train_index, test_index) in enumerate(kf.split(donor_X, donor_y), 1):
        X_train = donor_X[train_index]
        y_train = donor_y[train_index]  
        X_test = donor_X[test_index]
        y_test = donor_y[test_index]
        
        X_train_resampled, y_train_resampled = sme.fit_resample(X_train, y_train)
        
        classifier.fit(X_train_resampled, y_train_resampled )
        y_pred = classifier.predict(X_test)
        y_pred_train = classifier.predict(X_train)
        
        scores['Recall Score'].append(recall_score(y_test,  y_pred, average=average))
        scores['F1 Score'].append(f1_score(y_test,  y_pred, average=average))
        scores['Precision Score'].append(precision_score(y_test,  y_pred, average=average))
        scores['Accuracy Score'].append(accuracy_score(y_test,  y_pred))
        scores['Accuracy Score (train)'].append(accuracy_score(y_train,  y_pred_train))
        
        if plot:
            cm = confusion_matrix(y_test, y_pred)
            sns.heatmap(cm.T, square=True, annot=True, fmt='d', cbar=False, cmap=plt.cm.RdPu)
            plt.xlabel('true label')
            plt.ylabel('predicted label')
            plt.title('Confusion Matrix - Test Set')
            plt.show()
    return pd.DataFrame(scores)
```

```python
def get_binary_database(df):
    result = None
    for column in df.columns:
        te = TransactionEncoder()
        te_ary = te.fit(df[column].to_frame().values).transform(df[column].to_frame().values)
        binary_database = pd.DataFrame(te_ary, columns=te.columns_)
        binary_database = binary_database.add_suffix('_' + column)
        if result is None:
            result = binary_database
        else:
            #binary_database.add_suffix('_' + column)
            result = result.join(binary_database)
    return result
```
